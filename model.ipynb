Preparing the data

```python
df = pd.read_csv('./data/VAO_data.csv')
df.head()
```

```python
def clean_df(df):
    df = df.drop(['name', 'id', 'address', 'unom', 'district', 'area', 'col_775', 'col_764', 'col_766', 'col_1945_del'], axis=1)
    df = df.drop([0], axis=0)
    df = df.rename(columns={'col_758': 'projects-series-number', 'col_759': 'floors-amount', 'col_760': 'entrances-amount', 
                            'col_761': 'apartments-amount', 'col_762': 'whole-square', 'col_763': 'apartments-square',	
                            'col_769': 'materials', 'col_770': 'EMERGENCY-RATE', 'col_771': 'passengers-elevators', 
                            'col_772': 'cargo-elevators', 'col_781': 'roof-materials',
                            'col_2463': 'housing-stock', 'col_3163': 'status'})
    return df
# df = numpy.where(df['col_770'] == '22728486', 0, 1)
df = clean_df(df)
df
```

```python
train, valid, test = numpy.split(df.sample(frac=1), [int(0.6 * len(df)), int(0.8 * len(df))])
```

```python
def scale_dataset(data):
    data['EMERGENCY-RATE'] = (data['EMERGENCY-RATE'] == '22728488').astype(int)
    y = numpy.array(data[data.columns[7]].values)
    X = numpy.array(data.drop(data.columns[7], axis=1).values)
    data = numpy.array(numpy.hstack((X, numpy.reshape(y, (-1, 1)))))
    
    return data.astype(float), X.astype(float), y.astype(float)
```

```python
train_data, X_train, y_train = scale_dataset(train)
valid_data, X_valid, y_valid = scale_dataset(valid)
test_data, X_test, y_test = scale_dataset(train)
```

```python
def normalize(dataset):
    return tf.keras.utils.normalize(dataset)
```

```python
X_train = normalize(X_train)
X_test = normalize(X_test)
X_valid = normalize(X_valid)

X_train
```
